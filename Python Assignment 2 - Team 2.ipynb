## Survival rate of male and female passengers?

```python
male_survived = len(df[(df.Sex == 'male') & (df.Survived == 1)])
male_died = len(df[(df.Sex == 'male') & (df.Survived == 0)])

female_survived = len(df[(df.Sex == 'female') & (df.Survived == 1)])
female_died = len(df[(df.Sex == 'female') & (df.Survived == 0)])
```

```python
male_survival_rate = (male_survived / (male_survived + male_died)) * 100
female_survival_rate = (female_survived / (female_survived + female_died)) * 100

print("Survival rate of male passengers:", round(male_survival_rate,2),"%")
print("Survival rate of female passengers:", round(female_survival_rate,2),"%")
```

```python
p1_survived = len(df[(df.Pclass == 1) & (df.Survived == 1)])
p1_died = len(df[(df.Pclass == 1) & (df.Survived == 0)])

p2_survived = len(df[(df.Pclass == 2) & (df.Survived == 1)])
p2_died = len(df[(df.Pclass == 2) & (df.Survived == 0)])

p3_survived = len(df[(df.Pclass == 3) & (df.Survived == 1)])
p3_died = len(df[(df.Pclass == 3) & (df.Survived == 0)])


p1_survival_rate = p1_survived / (p1_survived + p1_died) * 100

p2_survival_rate = p2_survived / (p2_survived + p2_died) * 100

p3_survival_rate = p3_survived / (p3_survived + p3_died) * 100


print("Passenger class p1 survival rate:",round(p1_survival_rate,2), "%")
print("Passenger class p2 survival rate:",round(p2_survival_rate,2), "%")
print("Passenger class p3 survival rate:",round(p3_survival_rate,2), "%")

print("Class with Highest Survival rate",round(max(p1_survival_rate,p2_survival_rate,p3_survival_rate),2),"%")


```

## p1 has the highest survival rate of 62.96%

## Dictionary

```python
survival_rates_pclass = {
    'P1 Survival Rate': p1_survival_rate,
    'P2 Survival Rate': p2_survival_rate,
    'P3 Survival Rate': p3_survival_rate
}

print("Survival Rates by Ticket Class:")

for key, value in survival_rates_pclass.items():
    print('{}: {:.2f}'.format(key, value))
```

## Bar Chart

```python
p1_survived = len(df[(df.Pclass == 1) & (df.Survived == 1)])
p1_died = len(df[(df.Pclass == 1) & (df.Survived == 0)])

p2_survived = len(df[(df.Pclass == 2) & (df.Survived == 1)])
p2_died = len(df[(df.Pclass == 2) & (df.Survived == 0)])

p3_survived = len(df[(df.Pclass == 3) & (df.Survived == 1)])
p3_died = len(df[(df.Pclass == 3) & (df.Survived == 0)])


p1_survival_rate = p1_survived / (p1_survived + p1_died) * 100

p2_survival_rate = p2_survived / (p2_survived + p2_died) * 100

p3_survival_rate = p3_survived / (p3_survived + p3_died) * 100


a = [p1_survival_rate,p2_survival_rate,p3_survival_rate]
b = ['pclass=1','pclass=2','pclass=3']
plt.bar(b,a)
plt.title('Passenger Class vs Survival Rate')
plt.xlabel('Pclass')
plt.ylabel('Survival Rate')
plt.show()
```

## Scatterplot

```python
plt.scatter(df['Age'], df['Fare'], c=df['Survived'], cmap='coolwarm')

c_bar = plt.colorbar()
c_bar.set_label('Survival')

plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs Fare by Survival')
plt.show()
```

## Boxplot

```python
class1_fares = df[df['Pclass'] == 1]['Fare']
class2_fares = df[df['Pclass'] == 2]['Fare']
class3_fares = df[df['Pclass'] == 3]['Fare']
fares = [class1_fares, class2_fares, class3_fares]


fig, ax = plt.subplots()
ax.boxplot(fares)


ax.set_xticklabels(['Pclass = 1', 'Pclass = 2', 'Pclass = 3'])
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Fare')
ax.set_title('Distribution of Passenger Fares by Passenger Class')


plt.show()
```
